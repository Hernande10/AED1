```python
import pandas as pd
```

# Iniciando ETL (Extract, Transform and Load)

### Obtendo dados

```python
dataframe =pd.read_csv("http://dados.recife.pe.gov.br/dataset/4d3a3b39-9ea9-46ed-bf21-a2670de519c1/resource/7c613836-9edd-4c0f-bc72-495008dd29c3/download/info_escolas_2023_27122023.csv",
                       sep=";",parse_dates=[ 'data_visita'], dayfirst=True)
dataframe.head() # Por default o retorno é os 5 primeiros registros.
dataframe.head(15) #Os primeiros 15
dataframe.tail(6) #Os ultimos 6
```

### Removendos as colunas desnecessárias

```python
df = dataframe.drop ((['rpa','tipo_cod','cod_escola']), axis=1) #axis=1 indica que removemos as colunas
df.head()
```

### Vizualizando os tipos de dados

```python
df.dtypes #dtypes é uma propiedade e por isso não tem ()
```

### Visualizando apenas coluna visita

```python
df.data_visita
```

## Analise descritiva dos dados

```python
### Medidas de tendência central
```

```python
# Calcular a média da data da visita (Pode ser util para entender as visitas ao longo do ano)
df.data_visita.mean()
```

```python
#Vamos calcular a mediana da data de visita
#(Geralmente é util para identificar o "meio" das visitas, caso elas sejam assimétricas)

```

```python
df.data_visita.median()
```

```python
#Obtendo a menor data
df['data_visita'].max()
df['data_visita'].min()
```

## Medidas de dispersão

```python
df.describe() # Visualiza um resumo descritivo da todas as colunas numéricas do dataframe.
```

### Distribuição de frequencia

```python
df['tipo'].value_counts()
```

```python
df['tipo'].value_counts().plot(kind='bar')
```

```python
### Plotando os dados com Pyplot
```

```python
import matplotlib.pyplot as plt
#Contando a fraquencia dos tipos de escola
contagem_tipos = df.tipo.value_counts()

#Criando o gráfico de barra com personalizações
contagem_tipos.plot(kind='bar', color= 'blue')
#Adicionar personalizações com pyplot
plt.title('Distribuição das escolas de recife por tipo')
plt.xlabel('Tipo de escola')
plt.ylabel('Total de escola')
plt.xticks(rotation=45) #rotacionar os nomes dos eixos
plt.grid(axis='y')
plt.show()
```

```python
# Criando um gráfico de pizza
df.tipo.value_counts().plot(kind='pie', title='Distribuição das escolas por tipo', autopct='%1.1f%%')
```

## Meses com mais visitas

```python
linhas_com_datavazia = df.data_visita.isnull()
df_filtrado = df[~linhas_com_datavazia]
df_filtrado.head()
```

```python
media_visita = df_filtrado.data_visita.mean()
media_visita_mes = media_visita.month
print(f'Média do mês das visitas {media_visita_mes}')
```

```python
df_filtrado['data_visita'].dt.month.value_counts().plot(kind= 'hist', title='Distibuição Mensal das visitas')
```
